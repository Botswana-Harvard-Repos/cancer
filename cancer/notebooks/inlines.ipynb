```python
# Export CRFS
from django.db.models import get_model
crf_list = ['chemoMedplan', 'radiationtreatmentrecord', 'haartmedrecord']
for model_name in crf_list:
    cls = get_model('cancer_subject', model_name)
    objs = cls.objects.all()
    crf_data = []
    crf_file_path = data_path + model_name + '.csv'
    for obj in objs:
        data = obj.__dict__
        crf_data.append(data)
    df_crf = pd.DataFrame(crf_data)
    df_crf.to_csv(crf_file_path, sep='|', encoding='utf-8')
```

```python
# Import CRFs
import csv
from decimal import Decimal
from django.apps import apps as django_apps
from cancer_subject.models import SubjectVisit
from django.core.handlers import exception
from django.db.utils import DataError
from django.db.utils import IntegrityError
from builtins import ValueError
crf_list = ['chemomedplan', 'radiationtreatmentrecord', 'haartmedrecord']
data_path = '/Users/coulsonkgathi/source/data_files/20190522/inlines/'

problem = []
for crf_name in crf_list:
    fname = data_path + crf_name + '.csv'
    with open(fname, encoding="ISO-8859-1") as csv_file:
        csv_reader = csv.reader(csv_file, delimiter='|')
        header = next(csv_reader)
        del header[0]
        created = 0
        already_exists = 0
        count = 0
        for line in csv_reader:
            del line[0]
            if len(line) == len(header):
                data = dict(zip(header, line))
            else:
                print(lines[count - 1])
                print(line)
                print(lines[count + 1])
                raise ValidationError('Line is not the right size')
            data.update(
            created=parser.parse(data.get('created')).replace(tzinfo=tz),
            id=uuid.UUID(data.get('id')))
            if data['modified']:
                data.update(
                    modified=parser.parse(data.get('modified')).replace(tzinfo=tz))
            else:
                data.update(modified=None)
            if 'chemomedplan' == crf_name:
                if data['start_date']:
                    data.update(
                        start_date=parser.parse(data.get('start_date')).replace(tzinfo=tz).date())
                else:
                    data.update(start_date=None)
                if data['stop_date']:
                    data.update(
                        stop_date=parser.parse(data.get('stop_date')).replace(tzinfo=tz).date())
                else:
                    data.update(stop_date=None)
            if 'radiationtreatmentrecord' == crf_name:
                if data['start_date']:
                    data.update(
                        start_date=parser.parse(data.get('start_date')).replace(tzinfo=tz).date())
                else:
                    data.update(start_date=None)
                if data['end_date']:
                    data.update(
                        end_date=parser.parse(data.get('end_date')).replace(tzinfo=tz).date())
                else:
                    data.update(end_date=None)
                if data['dose_delivered']:
                    data.update(dose_delivered=int(float(data['dose_delivered'])))
                else:
                    data.update(dose_delivered=None)
                if data['dose_described']:
                    data.update(dose_described=int(float(data['dose_described'])))
                else:
                    data.update(dose_described=None)
                if data['fractions']:
                    data.update(fractions=int(float(data['fractions'])))
                else:
                    data.update(fractions=None)
                if data['dose_per_fraction']:
                    data.update(dose_per_fraction=int(float(data['dose_per_fraction'])))
                else:
                    data.update(dose_per_fraction=None)
            if 'haartmedrecord' == crf_name:
                if data['start_date']:
                    data.update(
                        start_date=parser.parse(data.get('start_date')).replace(tzinfo=tz).date())
                else:
                    data.update(start_date=None)
                if data['stop_date']:
                    data.update(
                        stop_date=parser.parse(data.get('stop_date')).replace(tzinfo=tz).date())
                else:
                    data.update(stop_date=None)
            del data['_state']
            crf_cls = django_apps.get_model('cancer_subject', crf_name)
            try:
                crf_cls.objects.get(id=data.get('id'))
            except crf_cls.DoesNotExist:
                try:
                    obj = crf_cls(**data)
                    obj.save_base(raw=True)
                except IntegrityError as e:
                    print(e)
                    problem.append({crf_name: data})
                else:
                    created += 1
            else:
                already_exists += 1
        print('**************************************************')
        print(f'Created: {created} for {crf_name}')
        print(f'Already Exists: {already_exists} for {crf_name}')
        print('###################################################')
print('Done woth all CRFS')
```
