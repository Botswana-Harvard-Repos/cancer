```python
import os, sys
import django

sys.path.append('../..') # add path to project root dir
os.environ["DJANGO_SETTINGS_MODULE"] = "cancer.settings"

# for more sophisticated setups, if you need to change connection settings (e.g. when using django-environ):
#os.environ["DATABASE_URL"] = "postgres://myuser:mypassword@localhost:54324/mydb"

# Connect to Django ORM
django.setup()
```

```python
from cancer_subject.models import Appointment
from django.core.exceptions import MultipleObjectsReturned
myapps = Appointment.objects.all()
app_updated = 0
duplicates = []
for ap in myapps:
    try:
        app = Appointment.objects.using('live').get(visit_code=ap.visit_code, subject_identifier=ap.subject_identifier)
    except Appointment.DoesNotExist:
        if not ap.subject_identifier[:5] == '045-4':
            print('WHy is it missing', ap.visit_code, ap.subject_identifier)
    except MultipleObjectsReturned:
        duplicates.append([ap.visit_code, ap.subject_identifier])
    else:
        print(app)
        app.created = ap.created
        app.modified = ap.modified
        app.appt_datetime = ap.appt_datetime
        app.appt_status = ap.appt_status
        app.appt_close_datetime = ap.appt_close_datetime
        app.save_base(raw=True, using='live')
        app_updated += 0
print(app_updated)
```
